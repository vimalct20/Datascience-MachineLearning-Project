```python
df = pd.read_csv("weight-height.csv")
```

```python
min_limit = df["Height"].quantile(0.01)
max_limit = df["Height"].quantile(0.99)
```

# Trimming


```python
df_new1  = df[(df["Height"]<max_limit)&(df["Height"]>min_limit)]
```

```python
sns.distplot(df_new1["Height"])
```

```python
df_new1.shape()
```

```python
df_new= np.where(df["Height"]>max_limit,max_limit,np.where(df["Height"]<min_limit,min_limit,df["Height"]))
```
