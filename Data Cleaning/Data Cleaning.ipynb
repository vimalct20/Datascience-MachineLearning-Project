# Filling missing values with imputation

```python
df3 = df.copy()
```

```python
df3.head()
```

```python
mean_value = df3["reviews_per_month"].mean()
df3["reviews_per_month"].fillna(mean_value,inplace=True)
```

```python
mean_value
```

```python
df3.isna().sum()
```

# Filling categorical Data

```python
df4 = df.copy()
```

```python
df4.nunique()
```

```python
df4["last_review"].fillna(df4["last_review"].value_counts().index[0],inplace=True)
```

```python
df4.isna().sum()
```
