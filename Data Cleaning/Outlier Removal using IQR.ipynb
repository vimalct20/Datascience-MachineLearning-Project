```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("scholarship.csv")
```

```python
data.head()
```

```python
sns.distplot(data["scholarship_exam_marks"])
```

```python
sns.boxplot(data["scholarship_exam_marks"])
```

```python
Q1 = data["scholarship_exam_marks"].quantile(0.25)
Q3 = data["scholarship_exam_marks"].quantile(0.75)
```

```python
Q1
```

```python
Q3
```

```python
data.describe()
```

```python
IQR= Q3-Q1
```

```python
mini = Q1 - 1.5*IQR
maxi = Q3 + 1.5*IQR
```

```python
mini
```

```python
maxi
```

```python
data[(data["scholarship_exam_marks"] < mini) | (data["scholarship_exam_marks"]>maxi)]
```

# Trimming

```python
new_data =  data[(data["scholarship_exam_marks"] > mini) & (data["scholarship_exam_marks"]<maxi)]
```

```python
new_data
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(data["scholarship_exam_marks"])
plt.subplot(2,2,2)
sns.distplot(new_data["scholarship_exam_marks"])
plt.subplot(2,2,3)
sns.boxplot(data["scholarship_exam_marks"])
plt.subplot(2,2,4)
sns.boxplot(new_data["scholarship_exam_marks"])
```

# Capping 

```python
data["scholarship_exam_marks"]= np.where(data["scholarship_exam_marks"]>maxi,maxi,np.where(data["scholarship_exam_marks"]<mini,mini,data["scholarship_exam_marks"]))
```
